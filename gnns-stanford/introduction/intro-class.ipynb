# Por que Grafos?
Grafos são uma linguagem geral apra descrever e analisar entidades com relações/interações. Domínios complexos tem uma estrutura relacional rifca, que pode ser representada como um grafo relacional, ao modelar explicitamente as relações, atingimos melhor performance

## Por que Graph Deep Learning é difícil?
Networks são complexas:
- Tamanho arbitrário e estrutura topologica complexa (por exemplo: não possuem localidade espacial como grades)
- Não possui ordenação de nós (nodes) fixa ou ponto de referência
- Features multimodais

## Como construir um grafo:
- O que são nós (nodes)?
    - nós = objetos do sistema (ex: partículas, usuários, papers) que aprendem representações usando informação da vizinhança no grafo.
- o que são arestas (edges)?
    - As edges definem quais nós trocam informação no processo de message passing.

A escolha da representação de rede correta de um certo domínio/problema determina nossa habilidade de usar redes com sucesso:
- Às vezes, tem uma única, não-ambígua representação
- Em outros casos, a não é de qualquer forma única
- A forma que você conecta links determina a questão que você estuda.

Há também considerações importantes, weights, properties, types e attributes:
- Weights: Pesos são valores numéricos associados às arestas que indicam a intensidade, distância ou importância da relação entre dois nós. Se não houver peso, o grafo é não ponderado (todas as arestas valem 1).

- Properties: grau, matriz de adjacência simétrica, conectividade e caminhos.

- Types: Types indicam a categoria de nós ou arestas. Exemplo: Usuário, produto, paper, autor. Tipos de arestas: _comprou_, _citou_, _segue_

- Attributes: Attributes são os dados ou características associados a nós ou arestas. Formalmente, são vetores de features. Exemplos: energia de uma partícula, idade de um usuário, embedding de um documento




## Bipartite graph
É um grafo que os nodes podem ser dividos em dois sets desconexos **_U_** e **_V_** tanto que cada link conecta a um node em **_U_** para um em **_V_**. Isto é, U e V são sets independentes.

Exemplos:
- Autores-para-Papers (eles autoraram)
- Atores-para-filmes (eles apareceram em)
- Usuários-para-filmes (eles deram nota)
- Receitas-para-ingredientes (eles contém)

"Redes dobradas":
No contexto de teoria de grafos / network science, significa colapsar um grafo bipartido em um grafo de um único tipo de nó, criando conexões entre nós que compartilham vizinhos.

## Aplicações de Graph ML
Em Graph Machine Learning, as tarefas podem ser definidas de acordo com o nível estrutural do grafo em que a previsão é feita. No nível de nó (node-level), o objetivo é prever propriedades de cada nó individualmente, utilizando suas features e a informação agregada de seus vizinhos. No nível de aresta (edge-level), o modelo busca prever relações entre pares de nós, como a existência, tipo ou peso de uma conexão. Já no nível de subgrafo ou comunidade (subgraph-level), a tarefa envolve identificar ou classificar grupos de nós que formam estruturas locais dentro do grafo. Por fim, no nível de grafo (graph-level), o modelo produz uma saída única para o grafo inteiro, sendo comum em problemas como classificação de grafos ou geração de novas estruturas. Essas diferentes escalas definem o tipo de informação que o modelo deve aprender a representar.

### Node-level network structure
![Node-level network structure](images/node_level_ns.png)

- Node degree: É a métrica mais simples. O grau indica quantas conexões um nó possui. Nós com grau alto costumam ser mais conectados e potencialmente mais influentes na rede. No grafo da figura, por exemplo, o nó D tem várias conexões (B, C, E, H), portanto possui grau alto.

- Node importance & position: Além do número de conexões, também importa onde o nó está localizado na rede. Algumas métricas capturam isso. Uma delas é betweenness centrality, que mede quantos caminhos mínimos entre pares de nós passam por um determinado nó. Se muitos caminhos passam por ele, o nó funciona como uma ponte estrutural na rede. Outra métrica é a distância média para os outros nós, que indica quão central ou periférico um nó está no grafo.

- Substructures around the node: Também é importante observar os padrões locais de conexão. Um nó pode fazer parte de estruturas como triângulos, comunidades ou hubs locais. Por exemplo, na figura, os nós A, B e C formam um triângulo, indicando um pequeno grupo altamente conectado.
